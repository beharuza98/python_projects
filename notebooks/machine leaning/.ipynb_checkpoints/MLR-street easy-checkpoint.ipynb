StreetEasy is New York City’s leading real estate marketplace — from studios to high-rises, Brooklyn Heights to Harlem.

We have partnered with the StreetEasy Research team for the Multiple Linear Regression (MLR) lesson, and you will be working with a .csv file that contains a sample of 5,000 rentals listings in Manhattan, Brooklyn, and Queens. You’ll find the correlations between several features and the rent, build/evaluate a MLR model

## Data Investigation

```python
# Basic imports related to analysis
import pandas as pd
import numpy as np

# Imports related to visualisation
import matplotlib.pyplot as plt

# Imports related to preprocessing
from sklearn.preprocessing import StandardScaler 

# Imports related to ml
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
streeteasy = pd.read_csv("https://raw.githubusercontent.com/sonnynomnom/Codecademy-Machine-Learning-Fundamentals/master/StreetEasy/queens.csv")

df = pd.DataFrame(streeteasy)

print(df.head())
```

```python
x = df[['bedrooms', 'bathrooms', 'size_sqft', 'min_to_subway', 'floor', 'building_age_yrs', 'no_fee', 'has_roofdeck', 'has_washer_dryer', 'has_doorman', 'has_elevator', 'has_dishwasher', 'has_patio', 'has_gym']]

y = df[['rent']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2,random_state=6)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## creating MLR models

```python
mlr=LinearRegression()
mlr.fit(x_train, y_train)
y_predict=mlr.predict(x_test)
```

```python
plt.scatter(y_test, y_predict, alpha=0.4)
plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Actual Rent vs Predicted Rent")
plt.show()
```

```python
print(mlr.coef_)
```

## correlation

```python
plt.scatter(df[['size_sqft']], df[['rent']], alpha=0.4)
plt.scatter(df[['min_to_subway']], df[['rent']], alpha=0.4)
plt.show()
```

## Residual analysis

```python
print("Train score:") #to find the mean squared error regression loss for the training set.
print(mlr.score(x_train, y_train))
print("Test score:")#to find the mean squared error regression loss for the testing set.
print(mlr.score(x_test, y_test))
residuals = y_predict - y_test
 
plt.scatter(y_predict, residuals, alpha=0.4)
plt.title('Residual Analysis')
 
plt.show()
```

```python
plt.scatter(y_test, y_predict)
plt.plot(range(20000), range(20000))

plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Actual Rent vs Predicted Rent")

plt.show()
```
